```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf

%matplotlib inline
```

```python
days = 20
tickers = {
    'Bitcoin': 'BTC-USD',
    'Ethereum': 'ETH-USD',
    'Cardano': 'ADA-USD',
    'Dogecoin': 'DOGE-USD',
    'Tesla': 'TSLA',
    'Nvdia': 'NVDA',
    'Facebook': 'FB',
    'Alphabet': 'GOOG',
    'Twitter': 'TWTR',
    'Apple': 'AAPL'
}
def get_data(days, tickers):
    df = pd.DataFrame()
    for company in tickers.keys():
        ticker = yf.Ticker(tickers[company])
        hist = ticker.history(period=f'{days}d')
        hist.index = hist.index.strftime('%d %B %Y')
        hist = hist[['Close']]
        hist.columns = [company]
        hist = hist.T
        hist.index.name = 'Name'
        df = pd.concat([df, hist])
    return df
```

```python
companies = ['Bitcoin', 'Ethereum']
data = df.loc[companies]
```

```python
data = data.T.reset_index()
```

```python
data.head()
```

```python
data = pd.melt(data, id_vars=['Date']).rename(
    columns={'value': 'Prices(USD)'}
)
data
```

```python
chart = (
    alt.Chart(data)
    .mark_line(opacity=0.8, clip=True)
    .encode(
    x='Date:T',
    y=alt.Y('Prices(USD):Q', stack=None, scale=alt.Scale(domain=[1000, 4000])),
    color='Name:N'    
    )
)
```
